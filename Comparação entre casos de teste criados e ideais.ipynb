```python
import matplotlib.pyplot as plt
import numpy as np

# Dados de exemplo
cadastros = ['Empresa', 'Usuário', 'Escritório', 'Revenda']
casos_criados = [10, 15, 20, 5]  # Casos de teste criados
casos_ideais = [15, 9, 35, 2]    # Casos de teste ideais

# Calculando o Índice de Completude
completude = [(criado / ideal) * 100 for criado, ideal in zip(casos_criados, casos_ideais)]

# Definindo a largura das barras
bar_width = 0.35
index = np.arange(len(cadastros))

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))

# Barras para Casos Criados e Casos Ideais
plt.bar(index - bar_width/2, casos_criados, bar_width, label='Casos de Teste Criados', color='skyblue')
plt.bar(index + bar_width/2, casos_ideais, bar_width, label='Casos de Teste Ideais', color='lightgreen')

# Adicionando título e rótulos
plt.title('Comparação entre Casos de Teste Criados e Casos de Teste Ideais', fontsize=14)
plt.xlabel('Cadastro', fontsize=12)
plt.ylabel('Quantidade de Casos de Teste', fontsize=12)
plt.xticks(index, cadastros)

# Adicionando a legenda
plt.legend()

# Exibindo o gráfico
plt.show()
```
